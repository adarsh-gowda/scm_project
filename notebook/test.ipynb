```python
s='madam'
i,j=0,len(s)-1
is_palindrome= True
while(i<j):
    if s[i] != s[j]:
        is_palindrome=False
        break
    i+=1
    j-=1
print("Polindrome" if is_palindrome else "Not a polindrom" )
```

```python
def is_prime(n):
    if n<=1:
        return False
    for i in range(2,int(n**0.5)+1):
        if n%i ==0:
            return False
    return True
def prime_lst(n):
    prime=[]
    for i in range(2,n+1):
        if is_prime(i):
            prime.append(i)
    return prime
num=int(input("Enter a number:"))
print(num,"is prime" if is_prime(num) else "is not prime")
print(f"all prime num upto {num}:",prime_lst(num))

```
